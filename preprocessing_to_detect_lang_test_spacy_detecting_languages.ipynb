```python
import pandas as pd
import spacy

nlp = spacy.load('en_core_web_sm')

def detect_language_spacy(text):
    try:
        doc = nlp(text)
        detected_language = doc.lang_
        return detected_language
    except Exception as e:
        print("An error occurred:", e)
        return None


depression_dataset['language'] = depression_dataset['clean_text'].apply(detect_language_spacy)

language_distribution = depression_dataset['language'].value_counts()

print(language_distribution)
```
