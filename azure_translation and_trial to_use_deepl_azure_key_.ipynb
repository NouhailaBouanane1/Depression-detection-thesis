```python
#Removing Punctuation & Special Characters
import string
string.punctuation
#defining the function to remove punctuation
def remove_punctuation(post):
    punctuation_free = "".join([chars for chars in post if chars not in string.punctuation])
    return punctuation_free

#storing the puntuation free text
depression_dataset['clean_text'] = depression_dataset['clean_text'].apply(remove_punctuation)
```

```python
import deepl

auth_key = " "
translator = deepl.Translator(auth_key)

# Create a language detector
detector = LanguageDetectorBuilder.from_all_languages().build()

# Function to detect the language of a text
def detect_language(text):
    result = detector.detect_language_of(text)
    return result.name

# Function to translate non-English text to English using DeepL
def translate_to_english(text):
    translation = translator.translate_text(text, target_lang='EN-US')
    return translation

# Apply language detection and translation to each row in the DataFrame
for index, row in depression_dataset.iterrows():
    detected_language = detect_language(row['clean_text'])
    if detected_language != 'ENGLISH':
        translated_text = translate_to_english(row['clean_text'])
        depression_dataset.at[index,'translated_text'] = translated_text
    else:
        depression_dataset.at[index,'translated_text'] = row['clean_text']

    print(f"Text: {row['clean_text']}, Translated_text: {row['translated_text']} Language: {detected_language}, Depressed: {row['is_depression']}")
depression_dataset.to_csv('translated_reddit_posts.csv')
```

```python
import requests
import pandas as pd  # Assuming you're using pandas DataFrame
import uuid

subscription_key = " "
endpoint = "https://api.cognitive.microsofttranslator.com/"
location = "westeurope"



# Initialize an empty list to store translated texts
translated_texts = []

# Loop through each row in the dataset
for index, row in depression_dataset.iterrows():
    text_to_translate = row['clean_text']  # Assuming 'clean_text' is the column containing the text to be translated
    detected_language = row['language']  # Assuming 'language' is the column containing the detected language

    # Check if the detected language is not English
    if detected_language != 'ENGLISH':
        # Translate the text to English
        headers = {
            'Ocp-Apim-Subscription-Key': subscription_key,
            'Ocp-Apim-Subscription-Region': location,
            'Content-type': 'application/json',
            'X-ClientTraceId': str(uuid.uuid4())
        }
        body = [{'text': text_to_translate}]
        params = {'api-version': '3.0', 'to': 'en'}
        response = requests.post(f"{endpoint}/translate", params=params, headers=headers, json=body)
        translation_result = response.json()
        translated_text = translation_result[0]['translations'][0]['text'] if translation_result else text_to_translate
    else:
        # Keep the text unchanged if it's already in English
        translated_text = text_to_translate

    # Append the translated text to the list
    translated_texts.append(translated_text)

# Add the translated texts to the dataset
depression_dataset['translated_text'] = translated_texts

# Save the modified dataset with translations
depression_dataset.to_csv('translated_reddit_posts.csv', index=False)
```
