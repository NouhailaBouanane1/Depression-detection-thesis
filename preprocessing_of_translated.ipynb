To reorder the df

```python
translated_dep_df= translated_dataset.drop(['clean_text','language'],axis='columns')
```

```python
# Reorder the columns with 'is_depressed' at the end
column_order = ['translated_text', 'is_depression']
translated_dep_df = translated_dep_df[column_order]
translated_dep_df
```

Removing URLS

```python
import pandas as pd
import re

# Define a regex pattern to match URLs
url_pattern = re.compile(r'https?://\S+')

# Define a function to remove URLs from text
def remove_urls(text):
    return url_pattern.sub('', text)

translated_dep_df['translated_text'] = translated_dep_df['translated_text'].apply(remove_urls)
```

Removing Punctuation

```python
import string
string.punctuation
```

```python
#defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
translated_dep_df['translated_text']= translated_dep_df['translated_text'].apply(remove_punctuation)

translated_dep_df.head(1)
```

```python
# Check if there are any digits in the 'translated_text' column
contains_digits = translated_dep_df['translated_text'].str.contains(r'\d').any()
print("Contains digits:", contains_digits)
```
